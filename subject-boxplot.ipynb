```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df1 = pd.read_csv('./outputs/output_base.csv')
error1 = df1['error']
sid1 = df1['subject_id']
# error1 = error1[:143796]
# sid1 = sid1[:143796]
method1 = np.repeat('Baseline', len(sid1))

df2 = pd.read_csv('./outputs/output_delta.csv')
error2 = df2['error']
sid2 = df2['subject_id']
method2 = np.repeat('Personalization', len(sid2))

error = list(error1)+list(error2)
sid = list(sid1)+list(sid2)
method = list(method1)+list(method2)

id_set = list(set(sid))


num_select = 30

fig, axs = plt.subplots(2,2,figsize=(15,8), sharey=True)
for i in range(2):
    for j in range(2):
        # mask = [sid[i] in list(set(sid))[0:20] for i in range(len(sid))]
        mask = []
        select_ids = list(set(sid))[num_select*(2*i+j):num_select*(2*i+j+1)] 
        for k in range(len(sid)):
            if sid[k] in select_ids:
                mask.append(1)
            else:
                mask.append(0)
        mask = np.array(mask).astype(bool)
        
        data = {
            'subject_id': np.array(sid)[mask],
            'error': np.array(error)[mask],
            'Method': np.array(method)[mask]
        }
        
        df = pd.DataFrame(data)
        
        
        sns.boxplot(x='subject_id', y='error', hue='Method', showfliers=False, data=df, ax=axs[i,j])
        # sns.boxplot(x='subject_id', y='error', data=df, ax=axs[0,0])
        # sns.boxplot(x='subject_id', y='error', data=df, ax=axs[0,1])
        # sns.boxplot(x='subject_id', y='error', data=df, ax=axs[1,0])
        # sns.boxplot(x='subject_id', y='error', data=df, ax=axs[1,1])
        
        axs[i,j].set_xticks(np.arange(1,31,2))
        axs[i,j].set_xticklabels( [str(num_select*(2*i+j)+k) for k in np.arange(2,31,2)])
        # axs[0,0].set_xticklabels( [str(20*0+i+1) for i in range(20)])
        # axs[0,1].set_xticklabels( [str(20*1+i+1) for i in range(20)])
        # axs[1,0].set_xticklabels( [str(20*2+i+1) for i in range(20)])
        # axs[1,1].set_xticklabels( [str(20*3+i+1) for i in range(20)])

        axs[i,j].set_xlabel('Subject index')
        # axs[0,0].set_xlabel('Subject index')
        # axs[0,1].set_xlabel('Subject index')
        # axs[1,0].set_xlabel('Subject index')
        # axs[1,1].set_xlabel('Subject index')

        axs[i,j].set_ylabel('Error (cm)')
        # axs[0,0].set_ylabel('Error (cm)')
        # axs[0,1].set_ylabel('Error (cm)')
        # axs[1,0].set_ylabel('Error (cm)')
        # axs[1,1].set_ylabel('Error (cm)')


# plt.xlabel('Subject index')
# plt.ylabel('Error')
# plt.xticks(ticks=[i for i in range(20)], labels=[str(i+1) for i in range(20)])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

methods = ['CNN', 'SE', 'BAM', 'CBAM', 'ViT', 'AFF-Net']
group1 = [1.95, 1.95,1.94,1.93,1.68,1.74]
group2 = [1.66,1.67,1.60,1.60,1.42,1.38]
improvement = [(b-p)/b*100 for b,p in zip(group1, group2)]
x = np.arange(len(methods))
width = 0.35
# sns.set_style('whitegrid')

fig,ax1 = plt.subplots(figsize=(8,5))
bars1 = ax1.bar(x-width/2, group1, width, label='Baseline', color='royalblue')
bars2 = ax1.bar(x+width/2, group2, width, label='Personalization', color='lightcoral')

# ax.set_xlabel('Methods', fontsize=12)
ax1.set_ylabel('Error (cm)', fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(methods)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(x, improvement, marker='o', linestyle='-', color='g', label='Improvement (%)')
ax2.legend(loc='upper right')

plt.show()
```
